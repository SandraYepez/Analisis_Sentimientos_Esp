```python
import spacy
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from concurrent.futures import ThreadPoolExecutor
from tqdm import tqdm
import time

# Descargar léxicos necesarios
nltk.download('vader_lexicon')

# Cargar el modelo de lenguaje en español de spaCy
nlp = spacy.load("es_core_news_sm")

# Crear una instancia del analizador de sentimientos VADER
sia = SentimentIntensityAnalyzer()

def preprocess_text(text):
    # Procesar el texto con spaCy
    doc = nlp(text)
    # Filtrar las palabras que no son stop words ni signos de puntuación
    tokens = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]
    # Unir los tokens de nuevo en un texto
    preprocessed_text = " ".join(tokens)
    return preprocessed_text

# Función para aplicar el preprocesamiento en paralelo
def preprocess_parallel(column):
    with ThreadPoolExecutor(max_workers=4) as executor:
        result = list(tqdm(executor.map(preprocess_text, column), total=len(column), desc="Procesando"))
    return result

# Función para realizar el análisis de sentimientos
def analyze_sentiment(tweet):
    sentiment = sia.polarity_scores(tweet)
    if sentiment['compound'] >= 0.05:
        return 'Positivo'
    elif sentiment['compound'] <= -0.05:
        return 'Negativo'
    else:
        return 'Neutral'

# Función para aplicar el análisis de sentimientos en paralelo
def analyze_sentiment_parallel(column):
    with ThreadPoolExecutor(max_workers=4) as executor:
        result = list(tqdm(executor.map(analyze_sentiment, column), total=len(column), desc="Analizando Sentimientos"))
    return result

# Medir el tiempo de inicio
start_time = time.time()

# Aplicar el preprocesamiento a la columna "Tweet Limpio"
tweets['Tweet_Procesado_spaCy'] = preprocess_parallel(tweets['Tweet Limpio'])

# Aplicar el análisis de sentimientos a la columna "Tweet Procesado spaCy"
tweets['Sentimiento_VADER'] = analyze_sentiment_parallel(tweets['Tweet_Procesado_spaCy'])

# Medir el tiempo de finalización
end_time = time.time()

# Calcular el tiempo total de ejecución
total_time = end_time - start_time

# Guardar el DataFrame resultante con los sentimientos analizados
tweets.to_csv('tweets_con_sentimientos_VADER_SPACY.csv', index=False)

# Muestra el tiempo total de ejecución
print(f"Tiempo total de ejecución: {total_time} segundos")
```

```python
# Obtener la lista de valores únicos en la columna 'Nombre BD'
categorias = tweets['Nombre BD'].unique()

# Crear un diccionario para almacenar los resultados por categoría
resultados_por_categoria = {}

# Calcular las diferencias porcentuales para cada categoría
for categoria in categorias:
    # Filtra los datos para la categoría actual
    categoria_data = tweets[tweets['Nombre BD'] == categoria]
    
    # Calcula los promedios de retweets por polaridad
    promedios = categoria_data.groupby('Sentimiento_VADER')['Retweet Count'].mean()
    
    # Calcula las diferencias absolutas entre polaridades
    diferencia_pos_neg = promedios['Positivo'] - promedios['Negativo']
    diferencia_pos_neu = promedios['Positivo'] - promedios['Neutral']
    diferencia_neu_neg = promedios['Neutral'] - promedios['Negativo']
    
    # Calcula las diferencias porcentuales
    porcentaje_pos_neg = (diferencia_pos_neg / promedios['Negativo']) * 100
    porcentaje_pos_neu = (diferencia_pos_neu / promedios['Neutral']) * 100
    porcentaje_neu_neg = (diferencia_neu_neg / promedios['Negativo']) * 100
    
    # Almacena los resultados en el diccionario
    resultados_por_categoria[categoria] = {
        'Dif. Porcentual POS-NEG': porcentaje_pos_neg,
        'Dif. Porcentual POS-NEU': porcentaje_pos_neu,
        'Dif. Porcentual NEU-NEG': porcentaje_neu_neg
    }

# Convierte el diccionario en un DataFrame
resultados_df_VADER = pd.DataFrame(resultados_por_categoria)

# Transpone el DataFrame para que las categorías sean filas y las métricas sean columnas
resultados_df_VADER = resultados_df_VADER.transpose()

# Guarda el DataFrame en un archivo CSV
resultados_df_VADER.to_csv("resultados_df_VADER_DAB.csv", index=False)

# Mostrar los resultados
print(resultados_df_VADER)
```

```python
from sklearn.metrics import confusion_matrix

# Datos de sentimientos de BERT y VADER
data = {
    'BERT': {
        'NEU': 297665,
        'POS': 92364,
        'NEG': 62248
    },
    'VADER': {
        'NEU': 413378,
        'POS': 30346,
        'NEG': 8553
    }
}

# Crear DataFrames de pandas a partir de los datos
df_bert = pd.DataFrame(data['BERT'], index=['Count'])
df_vader = pd.DataFrame(data['VADER'], index=['Count'])

# Transponer los DataFrames para facilitar la comparación
df_bert = df_bert.transpose()
df_vader = df_vader.transpose()

# 1. Comparación de Frecuencias
print("Comparación de Frecuencias:")
print(df_bert)
print(df_vader)

# 2. Porcentaje de Coincidencia
total_texts = df_bert['Count'].sum()
coincidence = (df_bert['Count'] == df_vader['Count']).sum() / total_texts * 100
print(f"Porcentaje de Coincidencia: {coincidence:.2f}%")

# 3. Matriz de Confusión
y_true = df_bert.index.values
y_pred = df_vader.index.values
conf_matrix = confusion_matrix(y_true, y_pred)
print("Matriz de Confusión:")
print(conf_matrix)

# 5. Análisis de Divergencias (ejemplo de diferencias entre BERT y VADER)
divergences = df_bert[df_bert['Count'] != df_vader['Count']]
print("Divergencias:")
print(divergences)

# 6. Análisis de Casos Extremos (ejemplo de caso donde BERT y VADER difieren significativamente)
extreme_case = df_bert[df_bert['Count'] - df_vader['Count'] > 100]
print("Caso Extremo:")
print(extreme_case)
```
